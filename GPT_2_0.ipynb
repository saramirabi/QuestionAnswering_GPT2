```python
import random
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the fine-tuned GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Function to generate unrelated answers
def generate_unrelated_answers():
    unrelated_phrases = [
        "The sky is blue.",
        "Bananas are yellow.",
        "The Earth orbits the sun.",
        "Water boils at 100 degrees Celsius.",
        "Roses are red, violets are blue."
    ]
    return [random.choice(unrelated_phrases) for _ in range(1)]

# Function to generate a related answer using fine-tuned GPT-2
def generate_related_answer(user_question):
    inputs = tokenizer.encode("Q: " + user_question, return_tensors="pt")
    answer = model.generate(inputs, max_length=200, num_return_sequences=1, temperature=0.9)
    decoded_answer = tokenizer.decode(answer[0], skip_special_tokens=True)
    return decoded_answer

# Get a random question from the user
user_question = input("Please ask a question: ")

# Generate 3 unrelated answers
unrelated_answers = generate_unrelated_answers()

# Generate 1 related answer using fine-tuned GPT-2
related_answer = generate_related_answer(user_question)

# Display the generated answers
print("\nUnrelated Answers:")
for i, answer in enumerate(unrelated_answers, start=1):
    print(f"Unrelated Answer {i}: {answer}")

print("\nRelated Answer:")
print(f"Question: {user_question}")
print(f"Generated Answer: {related_answer}")
```
